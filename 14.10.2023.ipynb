```python
import pandas as pd
import numpy as np
from random import*


dis1 = "Математика"
dis2 = "Физика"
dis3 = "История"
dis4 = "Русский язык"
dis5 = "Обществознание"
score1 = [3.6,4.2,3.3,4.6,4.2]
score2 = [2.5,3.7,4.3,3.9,4.1]
score3 = [2.9,3.3,4.5,2.8,4.2]
score4 = [4.9,3.8,4.6,4.1,3.4]
score5 = [3.1,3.9,4.2,2.8,3.5]

nd = {dis1:score1,dis2:score2,dis3:score3,dis4:score4,dis5:score5}


total_frame = pd.DataFrame(nd)

total_frame.index = ["Антонов Е.А.","Сидоров М.А","Суханов П.Е.","Петрова А.А.","Вологдин С.А."]

total_frame["Средний балл"] = ((total_frame[dis1]+total_frame[dis2]+total_frame[dis3]+total_frame[dis4]+total_frame[dis5])/5)
total_frame.insert(5,"Средний балл по гуманитарным наукам",(total_frame[dis3]+total_frame[dis4]+total_frame[dis5])/3)
total_frame.insert(6,"Средний балл по техническим наукам",(total_frame[dis1]+total_frame[dis2])/2)

np.random.seed(5)
b_random = pd.Series(np.random.normal(size=5),index=total_frame.index)
total_frame.loc[:,"Рандомный движ"] = b_random



dropped = total_frame.drop(["Антонов Е.А."],axis=0)
dropped

```
